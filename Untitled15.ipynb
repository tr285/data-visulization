```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ============================================================
# LOAD DATASET
# ============================================================

df = pd.read_csv("spotify_youtube_music_trends.csv")

print("=" * 70)
print("SPOTIFY & YOUTUBE MUSIC TRENDS ANALYSIS")
print("=" * 70)

print("\nOriginal Dataset Shape:", df.shape)


# ============================================================
# CLEAN COLUMN NAMES
# ============================================================

df.columns = (
    df.columns
    .str.strip()
    .str.lower()
)

print("\nDataset Columns:")
print(df.columns.tolist())


# ============================================================
# REMOVE DUPLICATES
# ============================================================

duplicate_count = df.duplicated().sum()

df = df.drop_duplicates()

print("\nDuplicates Removed:", duplicate_count)


# ============================================================
# CONVERT NUMERIC COLUMNS
# ============================================================

numeric_columns = [

    "release_year",

    "spotify_streams_millions",

    "youtube_views_millions",

    "youtube_likes_millions",

    "youtube_comments_thousands",

    "popularity_score",

    "danceability",

    "energy",

    "valence",

    "acousticness",

    "instrumentalness",

    "liveness",

    "speechiness",

    "tempo_bpm",

    "loudness_db",

    "duration_sec"
]


for column in numeric_columns:

    if column in df.columns:

        df[column] = pd.to_numeric(
            df[column],
            errors="coerce"
        )

        df[column] = df[column].fillna(
            df[column].median()
        )


# ============================================================
# CLEAN CATEGORICAL COLUMNS
# ============================================================

categorical_columns = df.select_dtypes(
    include="object"
).columns


for column in categorical_columns:

    if df[column].isnull().sum() > 0:

        mode_value = df[column].mode()

        if not mode_value.empty:

            df[column] = df[column].fillna(
                mode_value[0]
            )

        else:

            df[column] = df[column].fillna(
                "Unknown"
            )


print(
    "\nMissing Values After Cleaning:",
    df.isnull().sum().sum()
)


print(
    "\nCleaned Dataset Shape:",
    df.shape
)


# ============================================================
# PREPARE DATA
# ============================================================


top_spotify_artists = (

    df.groupby("artist")
    ["spotify_streams_millions"]

    .sum()

    .nlargest(10)

    .sort_values()

)


top_youtube_artists = (

    df.groupby("artist")
    ["youtube_views_millions"]

    .sum()

    .nlargest(10)

    .sort_values()

)


top_spotify_tracks = (

    df.groupby("track")
    ["spotify_streams_millions"]

    .max()

    .nlargest(10)

    .sort_values()

)


top_youtube_tracks = (

    df.groupby("track")
    ["youtube_views_millions"]

    .max()

    .nlargest(10)

    .sort_values()

)


top_liked_tracks = (

    df.groupby("track")
    ["youtube_likes_millions"]

    .max()

    .nlargest(10)

    .sort_values()

)


genre_distribution = (

    df["genre"]

    .value_counts()

    .head(10)

)


platform_distribution = (

    df["platform_origin"]

    .value_counts()

)


explicit_distribution = (

    df["explicit"]

    .astype(str)

    .value_counts()

)


audio_features = [

    "danceability",

    "energy",

    "valence",

    "acousticness",

    "instrumentalness",

    "liveness",

    "speechiness"
]


audio_average = (

    df[audio_features]

    .mean()

    .sort_values(ascending=False)

)


year_trend = (

    df.groupby("release_year")

    .agg(

        Spotify_Streams=(
            "spotify_streams_millions",
            "mean"
        ),

        YouTube_Views=(
            "youtube_views_millions",
            "mean"
        )

    )

    .sort_index()

)


# ============================================================
# CREATE PROFESSIONAL DASHBOARD
# ============================================================

fig, axes = plt.subplots(

    4,

    3,

    figsize=(24, 30)

)


fig.suptitle(

    "Spotify & YouTube Music Trends Analytics Dashboard",

    fontsize=28,

    fontweight="bold",

    y=0.995

)


# ============================================================
# GRAPH 1
# TOP SPOTIFY ARTISTS
# ============================================================

ax = axes[0, 0]


ax.barh(

    top_spotify_artists.index,

    top_spotify_artists.values

)


ax.set_title(

    "Top 10 Artists by Spotify Streams",

    fontsize=16,

    fontweight="bold",

    pad=15

)


ax.set_xlabel(

    "Total Spotify Streams (Millions)"

)


ax.set_ylabel("Artist")


ax.grid(

    axis="x",

    alpha=0.3

)


# ============================================================
# GRAPH 2
# TOP YOUTUBE ARTISTS
# ============================================================

ax = axes[0, 1]


ax.barh(

    top_youtube_artists.index,

    top_youtube_artists.values

)


ax.set_title(

    "Top 10 Artists by YouTube Views",

    fontsize=16,

    fontweight="bold",

    pad=15

)


ax.set_xlabel(

    "Total YouTube Views (Millions)"

)


ax.set_ylabel("Artist")


ax.grid(

    axis="x",

    alpha=0.3

)


# ============================================================
# GRAPH 3
# GENRE DISTRIBUTION
# ============================================================

ax = axes[0, 2]


ax.bar(

    genre_distribution.index,

    genre_distribution.values

)


ax.set_title(

    "Top 10 Genres by Number of Tracks",

    fontsize=16,

    fontweight="bold",

    pad=15

)


ax.set_xlabel("Genre")

ax.set_ylabel("Number of Tracks")


ax.tick_params(

    axis="x",

    rotation=45

)


ax.grid(

    axis="y",

    alpha=0.3

)


# ============================================================
# GRAPH 4
# TOP SPOTIFY TRACKS
# ============================================================

ax = axes[1, 0]


ax.barh(

    top_spotify_tracks.index,

    top_spotify_tracks.values

)


ax.set_title(

    "Top 10 Tracks by Spotify Streams",

    fontsize=16,

    fontweight="bold",

    pad=15

)


ax.set_xlabel(

    "Spotify Streams (Millions)"

)


ax.set_ylabel("Track")


ax.grid(

    axis="x",

    alpha=0.3

)


# ============================================================
# GRAPH 5
# TOP YOUTUBE TRACKS
# ============================================================

ax = axes[1, 1]


ax.barh(

    top_youtube_tracks.index,

    top_youtube_tracks.values

)


ax.set_title(

    "Top 10 Tracks by YouTube Views",

    fontsize=16,

    fontweight="bold",

    pad=15

)


ax.set_xlabel(

    "YouTube Views (Millions)"

)


ax.set_ylabel("Track")


ax.grid(

    axis="x",

    alpha=0.3

)


# ============================================================
# GRAPH 6
# PLATFORM DISTRIBUTION
# ============================================================

ax = axes[1, 2]


ax.pie(

    platform_distribution.values,

    labels=platform_distribution.index,

    autopct="%1.1f%%",

    startangle=90

)


ax.set_title(

    "Platform Origin Distribution",

    fontsize=16,

    fontweight="bold",

    pad=15

)


# ============================================================
# GRAPH 7
# SPOTIFY VS YOUTUBE
# ============================================================

ax = axes[2, 0]


ax.scatter(

    df["spotify_streams_millions"],

    df["youtube_views_millions"],

    alpha=0.5,

    s=30

)


ax.set_title(

    "Spotify Streams vs YouTube Views",

    fontsize=16,

    fontweight="bold",

    pad=15

)


ax.set_xlabel(

    "Spotify Streams (Millions)"

)


ax.set_ylabel(

    "YouTube Views (Millions)"

)


ax.grid(alpha=0.3)


# ============================================================
# GRAPH 8
# YOUTUBE VIEWS VS LIKES
# ============================================================

ax = axes[2, 1]


ax.scatter(

    df["youtube_views_millions"],

    df["youtube_likes_millions"],

    alpha=0.5,

    s=30

)


ax.set_title(

    "YouTube Views vs Likes",

    fontsize=16,

    fontweight="bold",

    pad=15

)


ax.set_xlabel(

    "YouTube Views (Millions)"

)


ax.set_ylabel(

    "YouTube Likes (Millions)"

)


ax.grid(alpha=0.3)


# ============================================================
# GRAPH 9
# AUDIO FEATURE PROFILE
# ============================================================

ax = axes[2, 2]


ax.bar(

    audio_average.index,

    audio_average.values

)


ax.set_title(

    "Average Audio Feature Profile",

    fontsize=16,

    fontweight="bold",

    pad=15

)


ax.set_xlabel("Audio Feature")

ax.set_ylabel("Average Value")


ax.tick_params(

    axis="x",

    rotation=50

)


ax.grid(

    axis="y",

    alpha=0.3

)


# ============================================================
# GRAPH 10
# RELEASE YEAR TREND
# ============================================================

ax = axes[3, 0]


ax.plot(

    year_trend.index,

    year_trend["Spotify_Streams"],

    marker="o",

    label="Spotify Streams"

)


ax.plot(

    year_trend.index,

    year_trend["YouTube_Views"],

    marker="o",

    label="YouTube Views"

)


ax.set_title(

    "Average Music Popularity by Release Year",

    fontsize=16,

    fontweight="bold",

    pad=15

)


ax.set_xlabel("Release Year")

ax.set_ylabel("Average Performance (Millions)")


ax.legend()


ax.grid(alpha=0.3)


# ============================================================
# GRAPH 11
# TOP LIKED TRACKS
# ============================================================

ax = axes[3, 1]


ax.barh(

    top_liked_tracks.index,

    top_liked_tracks.values

)


ax.set_title(

    "Top 10 Tracks by YouTube Likes",

    fontsize=16,

    fontweight="bold",

    pad=15

)


ax.set_xlabel(

    "YouTube Likes (Millions)"

)


ax.set_ylabel("Track")


ax.grid(

    axis="x",

    alpha=0.3

)


# ============================================================
# GRAPH 12
# POPULARITY SCORE DISTRIBUTION
# ============================================================

ax = axes[3, 2]


ax.hist(

    df["popularity_score"],

    bins=20,

    edgecolor="black"

)


ax.set_title(

    "Distribution of Popularity Scores",

    fontsize=16,

    fontweight="bold",

    pad=15

)


ax.set_xlabel("Popularity Score")

ax.set_ylabel("Number of Tracks")


ax.grid(

    axis="y",

    alpha=0.3

)


# ============================================================
# ADD PROFESSIONAL SPACING
# ============================================================

plt.subplots_adjust(

    left=0.07,

    right=0.97,

    top=0.95,

    bottom=0.05,

    wspace=0.40,

    hspace=0.55

)


# ============================================================
# SAVE DASHBOARD
# ============================================================

plt.savefig(

    "spotify_youtube_dashboard.png",

    dpi=300,

    bbox_inches="tight"

)


plt.show()


# ============================================================
# SPACE BETWEEN DASHBOARD AND NEXT GRAPH
# ============================================================

print("\n\n\n\n")


# ============================================================
# CORRELATION MATRIX
# ============================================================

correlation_columns = [

    "spotify_streams_millions",

    "youtube_views_millions",

    "youtube_likes_millions",

    "youtube_comments_thousands",

    "popularity_score",

    "danceability",

    "energy",

    "valence",

    "acousticness",

    "tempo_bpm"

]


correlation = (

    df[correlation_columns]

    .corr()

)


plt.figure(figsize=(15, 11))


image = plt.imshow(

    correlation,

    aspect="auto",

    vmin=-1,

    vmax=1

)


plt.colorbar(

    image,

    label="Correlation Coefficient"

)


plt.xticks(

    range(len(correlation_columns)),

    correlation_columns,

    rotation=50,

    ha="right"

)


plt.yticks(

    range(len(correlation_columns)),

    correlation_columns

)


for row in range(len(correlation_columns)):

    for column in range(len(correlation_columns)):

        plt.text(

            column,

            row,

            f"{correlation.iloc[row, column]:.2f}",

            ha="center",

            va="center",

            fontsize=9

        )


plt.title(

    "Correlation Matrix of Music Performance Metrics",

    fontsize=22,

    fontweight="bold",

    pad=25

)


plt.tight_layout(pad=4)


plt.savefig(

    "music_correlation_matrix.png",

    dpi=300,

    bbox_inches="tight"

)


plt.show()


# ============================================================
# SPACE
# ============================================================

print("\n\n\n\n")


# ============================================================
# AUTOMATIC INSIGHTS
# ============================================================

print("=" * 70)

print("KEY FINDINGS AND INSIGHTS")

print("=" * 70)


best_spotify_artist = (

    df.groupby("artist")
    ["spotify_streams_millions"]

    .sum()

    .idxmax()

)


best_youtube_artist = (

    df.groupby("artist")
    ["youtube_views_millions"]

    .sum()

    .idxmax()

)


best_spotify_track = (

    df.loc[

        df["spotify_streams_millions"].idxmax(),

        "track"

    ]

)


best_youtube_track = (

    df.loc[

        df["youtube_views_millions"].idxmax(),

        "track"

    ]

)


stream_view_correlation = (

    df["spotify_streams_millions"]

    .corr(

        df["youtube_views_millions"]

    )

)


views_likes_correlation = (

    df["youtube_views_millions"]

    .corr(

        df["youtube_likes_millions"]

    )

)


most_common_genre = (

    df["genre"]

    .value_counts()

    .idxmax()

)


most_popular_year = (

    df.groupby("release_year")
    ["popularity_score"]

    .mean()

    .idxmax()

)


print(

    f"\n1. Leading Spotify Artist: "

    f"{best_spotify_artist}"

)


print(

    f"\n2. Leading YouTube Artist: "

    f"{best_youtube_artist}"

)


print(

    f"\n3. Most Streamed Track: "

    f"{best_spotify_track}"

)


print(

    f"\n4. Most Viewed YouTube Track: "

    f"{best_youtube_track}"

)


print(

    f"\n5. Most Common Genre: "

    f"{most_common_genre}"

)


print(

    f"\n6. Release Year with Highest Average Popularity: "

    f"{int(most_popular_year)}"

)


print(

    f"\n7. Spotify Streams vs YouTube Views Correlation: "

    f"{stream_view_correlation:.2f}"

)


print(

    f"\n8. YouTube Views vs Likes Correlation: "

    f"{views_likes_correlation:.2f}"

)


# ============================================================
# SAVE CLEANED DATASET
# ============================================================

df.to_csv(

    "cleaned_spotify_youtube_music_trends.csv",

    index=False

)


print("\n" + "=" * 70)

print("ANALYSIS COMPLETED SUCCESSFULLY")

print("=" * 70)


print(

    "\nDashboard saved as: spotify_youtube_dashboard.png"

)


print(

    "Correlation Matrix saved as: music_correlation_matrix.png"

)


print(

    "Cleaned Dataset saved as: "

    "cleaned_spotify_youtube_music_trends.csv"

)
```
