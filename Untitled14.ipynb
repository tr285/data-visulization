```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# =========================
# 1. LOAD DATASET
# =========================

file_path = "city_day_aqi_india.csv"

df = pd.read_csv(file_path)

print("\nOriginal Dataset Shape:", df.shape)
print("\nFirst 5 Rows:")
print(df.head())

print("\nDataset Information:")
df.info()


# =========================
# 2. REMOVE DUPLICATES
# =========================

duplicates = df.duplicated().sum()

print("\nDuplicate Rows:", duplicates)

df = df.drop_duplicates()


# =========================
# 3. CONVERT DATE COLUMN
# =========================

df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

df = df.dropna(subset=["Date"])


# =========================
# 4. CHECK MISSING VALUES
# =========================

print("\nMissing Values Before Cleaning:")
print(df.isnull().sum())


# =========================
# 5. HANDLE MISSING VALUES
# =========================

numeric_columns = df.select_dtypes(include=np.number).columns

for column in numeric_columns:
    df[column] = df.groupby("City")[column].transform(
        lambda x: x.fillna(x.median())
    )

for column in numeric_columns:
    df[column] = df[column].fillna(df[column].median())

df["AQI_Bucket"] = df["AQI_Bucket"].fillna("Unknown")


# =========================
# 6. REMOVE INVALID VALUES
# =========================

pollution_columns = [
    "PM2.5",
    "PM10",
    "NO",
    "NO2",
    "NOx",
    "NH3",
    "CO",
    "SO2",
    "O3",
    "Benzene",
    "Toluene",
    "Xylene",
    "AQI"
]

existing_pollution_columns = [
    col for col in pollution_columns
    if col in df.columns
]

for column in existing_pollution_columns:
    df.loc[df[column] < 0, column] = np.nan

for column in existing_pollution_columns:
    df[column] = df[column].fillna(df[column].median())


# =========================
# 7. CREATE DATE FEATURES
# =========================

df["Year"] = df["Date"].dt.year

df["Month"] = df["Date"].dt.month

df["Month_Name"] = df["Date"].dt.month_name()

df["Day"] = df["Date"].dt.day


# =========================
# 8. SUMMARY STATISTICS
# =========================

print("\nCleaned Dataset Shape:", df.shape)

print("\nSummary Statistics:")

print(df.describe().round(2))


# =========================
# 9. MISSING VALUES GRAPH
# =========================

missing = df.isnull().sum()

missing = missing[missing > 0]

if len(missing) > 0:

    plt.figure(figsize=(12, 6))

    missing.sort_values().plot(kind="barh")

    plt.title("Missing Values After Cleaning")

    plt.xlabel("Missing Value Count")

    plt.ylabel("Columns")

    plt.tight_layout()

    plt.show()


# =========================
# 10. AQI HISTOGRAM
# =========================

plt.figure(figsize=(10, 6))

plt.hist(
    df["AQI"].dropna(),
    bins=30,
    edgecolor="black"
)

plt.title("Distribution of AQI")

plt.xlabel("AQI")

plt.ylabel("Frequency")

plt.tight_layout()

plt.show()


# =========================
# 11. AQI BOX PLOT
# =========================

plt.figure(figsize=(10, 5))

plt.boxplot(
    df["AQI"].dropna(),
    vert=False
)

plt.title("AQI Box Plot")

plt.xlabel("AQI")

plt.tight_layout()

plt.show()


# =========================
# 12. TOP 15 POLLUTED CITIES
# =========================

city_aqi = (
    df.groupby("City")["AQI"]
    .mean()
    .sort_values(ascending=False)
    .head(15)
)

plt.figure(figsize=(12, 7))

city_aqi.sort_values().plot(kind="barh")

plt.title("Top 15 Cities by Average AQI")

plt.xlabel("Average AQI")

plt.ylabel("City")

plt.tight_layout()

plt.show()


# =========================
# 13. CLEANEST 15 CITIES
# =========================

clean_cities = (
    df.groupby("City")["AQI"]
    .mean()
    .sort_values()
    .head(15)
)

plt.figure(figsize=(12, 7))

clean_cities.plot(kind="bar")

plt.title("Top 15 Cities with Lowest Average AQI")

plt.xlabel("City")

plt.ylabel("Average AQI")

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()


# =========================
# 14. AQI TREND OVER TIME
# =========================

daily_aqi = (
    df.groupby("Date")["AQI"]
    .mean()
)

plt.figure(figsize=(14, 6))

plt.plot(
    daily_aqi.index,
    daily_aqi.values
)

plt.title("Average AQI Trend Over Time")

plt.xlabel("Date")

plt.ylabel("Average AQI")

plt.grid(True)

plt.tight_layout()

plt.show()


# =========================
# 15. YEAR-WISE AQI TREND
# =========================

yearly_aqi = (
    df.groupby("Year")["AQI"]
    .mean()
)

plt.figure(figsize=(10, 6))

plt.plot(
    yearly_aqi.index,
    yearly_aqi.values,
    marker="o"
)

plt.title("Year-wise Average AQI")

plt.xlabel("Year")

plt.ylabel("Average AQI")

plt.grid(True)

plt.tight_layout()

plt.show()


# =========================
# 16. MONTHLY AQI TREND
# =========================

monthly_aqi = (
    df.groupby("Month")["AQI"]
    .mean()
)

plt.figure(figsize=(10, 6))

plt.plot(
    monthly_aqi.index,
    monthly_aqi.values,
    marker="o"
)

plt.title("Monthly Average AQI")

plt.xlabel("Month")

plt.ylabel("Average AQI")

plt.xticks(range(1, 13))

plt.grid(True)

plt.tight_layout()

plt.show()


# =========================
# 17. AQI BUCKET PIE CHART
# =========================

aqi_bucket_counts = df["AQI_Bucket"].value_counts()

plt.figure(figsize=(9, 9))

plt.pie(
    aqi_bucket_counts.values,
    labels=aqi_bucket_counts.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("AQI Category Distribution")

plt.tight_layout()

plt.show()


# =========================
# 18. AQI BUCKET BAR GRAPH
# =========================

plt.figure(figsize=(10, 6))

aqi_bucket_counts.plot(kind="bar")

plt.title("AQI Category Counts")

plt.xlabel("AQI Category")

plt.ylabel("Number of Records")

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()


# =========================
# 19. POLLUTANT HISTOGRAMS
# =========================

pollutants = [
    "PM2.5",
    "PM10",
    "NO2",
    "CO",
    "SO2",
    "O3"
]

available_pollutants = [
    column for column in pollutants
    if column in df.columns
]

for pollutant in available_pollutants:

    plt.figure(figsize=(10, 6))

    plt.hist(
        df[pollutant].dropna(),
        bins=30,
        edgecolor="black"
    )

    plt.title(f"{pollutant} Distribution")

    plt.xlabel(pollutant)

    plt.ylabel("Frequency")

    plt.tight_layout()

    plt.show()


# =========================
# 20. POLLUTANT BOX PLOTS
# =========================

for pollutant in available_pollutants:

    plt.figure(figsize=(10, 5))

    plt.boxplot(
        df[pollutant].dropna(),
        vert=False
    )

    plt.title(f"{pollutant} Box Plot")

    plt.xlabel(pollutant)

    plt.tight_layout()

    plt.show()


# =========================
# 21. SCATTER PLOT
# PM2.5 VS AQI
# =========================

if "PM2.5" in df.columns:

    sample_df = df[["PM2.5", "AQI"]].dropna()

    if len(sample_df) > 5000:
        sample_df = sample_df.sample(
            5000,
            random_state=42
        )

    plt.figure(figsize=(10, 6))

    plt.scatter(
        sample_df["PM2.5"],
        sample_df["AQI"],
        alpha=0.5
    )

    plt.title("PM2.5 vs AQI")

    plt.xlabel("PM2.5")

    plt.ylabel("AQI")

    plt.grid(True)

    plt.tight_layout()

    plt.show()


# =========================
# 22. SCATTER PLOT
# PM10 VS AQI
# =========================

if "PM10" in df.columns:

    sample_df = df[["PM10", "AQI"]].dropna()

    if len(sample_df) > 5000:
        sample_df = sample_df.sample(
            5000,
            random_state=42
        )

    plt.figure(figsize=(10, 6))

    plt.scatter(
        sample_df["PM10"],
        sample_df["AQI"],
        alpha=0.5
    )

    plt.title("PM10 vs AQI")

    plt.xlabel("PM10")

    plt.ylabel("AQI")

    plt.grid(True)

    plt.tight_layout()

    plt.show()


# =========================
# 23. AREA CHART
# =========================

yearly_pollution = (
    df.groupby("Year")[available_pollutants]
    .mean()
)

if len(available_pollutants) > 0:

    plt.figure(figsize=(14, 7))

    plt.stackplot(
        yearly_pollution.index,
        yearly_pollution.T.values,
        labels=available_pollutants
    )

    plt.title("Year-wise Pollution Area Chart")

    plt.xlabel("Year")

    plt.ylabel("Average Pollutant Concentration")

    plt.legend(loc="upper left")

    plt.tight_layout()

    plt.show()


# =========================
# 24. STACKED BAR CHART
# =========================

top_cities = (
    df.groupby("City")["AQI"]
    .mean()
    .nlargest(10)
    .index
)

stacked_data = (
    df[df["City"].isin(top_cities)]
    .groupby(["City", "AQI_Bucket"])
    .size()
    .unstack(fill_value=0)
)

plt.figure(figsize=(14, 8))

stacked_data.plot(
    kind="bar",
    stacked=True,
    figsize=(14, 8)
)

plt.title("AQI Categories in Top Polluted Cities")

plt.xlabel("City")

plt.ylabel("Number of Records")

plt.xticks(rotation=45)

plt.legend(
    title="AQI Category",
    bbox_to_anchor=(1.05, 1)
)

plt.tight_layout()

plt.show()


# =========================
# 25. CORRELATION HEATMAP
# WITHOUT SEABORN
# =========================

correlation_columns = existing_pollution_columns

correlation_matrix = (
    df[correlation_columns]
    .corr()
)

plt.figure(figsize=(14, 10))

plt.imshow(
    correlation_matrix,
    aspect="auto"
)

plt.colorbar(label="Correlation")

plt.xticks(
    range(len(correlation_matrix.columns)),
    correlation_matrix.columns,
    rotation=90
)

plt.yticks(
    range(len(correlation_matrix.columns)),
    correlation_matrix.columns
)

plt.title("Pollutant Correlation Heatmap")

plt.tight_layout()

plt.show()


# =========================
# 26. CITY VS YEAR AQI HEATMAP
# =========================

top_20_cities = (
    df.groupby("City")["AQI"]
    .mean()
    .nlargest(20)
    .index
)

city_year_data = (
    df[df["City"].isin(top_20_cities)]
    .pivot_table(
        values="AQI",
        index="City",
        columns="Year",
        aggfunc="mean"
    )
)

plt.figure(figsize=(14, 9))

plt.imshow(
    city_year_data,
    aspect="auto"
)

plt.colorbar(label="Average AQI")

plt.xticks(
    range(len(city_year_data.columns)),
    city_year_data.columns
)

plt.yticks(
    range(len(city_year_data.index)),
    city_year_data.index
)

plt.title("City-wise Year-wise AQI Heatmap")

plt.xlabel("Year")

plt.ylabel("City")

plt.tight_layout()

plt.show()


# =========================
# 27. CITY-WISE AQI SUMMARY
# =========================

city_summary = (
    df.groupby("City")
    .agg(
        Average_AQI=("AQI", "mean"),
        Maximum_AQI=("AQI", "max"),
        Minimum_AQI=("AQI", "min"),
        Records=("AQI", "count")
    )
    .sort_values(
        "Average_AQI",
        ascending=False
    )
)

print("\nCity-wise AQI Summary:")

print(city_summary.round(2))


# =========================
# 28. FIND MOST POLLUTED CITY
# =========================

most_polluted_city = city_summary.index[0]

print(
    "\nMost Polluted City:",
    most_polluted_city
)

print(
    "Average AQI:",
    round(
        city_summary.iloc[0]["Average_AQI"],
        2
    )
)


# =========================
# 29. FIND CLEANEST CITY
# =========================

cleanest_city = (
    city_summary
    .sort_values("Average_AQI")
    .index[0]
)

print(
    "\nCleanest City:",
    cleanest_city
)


# =========================
# 30. SAVE CLEANED DATASET
# =========================

output_file = "cleaned_city_day_aqi_india.csv"

df.to_csv(
    output_file,
    index=False
)

city_summary.to_csv(
    "city_aqi_summary.csv"
)

print("\nDATA CLEANING COMPLETED")

print("EDA COMPLETED")

print("ALL GRAPHS GENERATED")

print("Cleaned Dataset Saved As:", output_file)

print("City Summary Saved As: city_aqi_summary.csv")
```
