```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

# -------------------------------
# Configuration
# -------------------------------
INPUT_FILE = "spotify_tracks_dataset.csv"
OUTPUT_FOLDER = "Spotify_Dashboard_Output"

os.makedirs(OUTPUT_FOLDER, exist_ok=True)

# -------------------------------
# Load Dataset
# -------------------------------
df = pd.read_csv(INPUT_FILE)

# -------------------------------
# Clean Column Names
# -------------------------------
df.columns = (
    df.columns.str.strip()
              .str.lower()
              .str.replace(" ", "_")
              .str.replace("-", "_")
)

print("Columns:")
print(df.columns.tolist())

# -------------------------------
# Remove Duplicates
# -------------------------------
df.drop_duplicates(inplace=True)

# -------------------------------
# Handle Missing Values
# -------------------------------
for col in df.columns:
    if df[col].dtype == "object":
        df[col] = df[col].fillna("Unknown")
    else:
        df[col] = df[col].fillna(df[col].median())

# -------------------------------
# Convert Numeric Columns
# -------------------------------
for col in df.columns:
    if col not in df.select_dtypes(include="object").columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")

df.dropna(inplace=True)

# -------------------------------
# Save Cleaned Dataset
# -------------------------------
clean_file = os.path.join(OUTPUT_FOLDER, "spotify_cleaned.csv")
df.to_csv(clean_file, index=False)

summary_file = os.path.join(OUTPUT_FOLDER, "summary_statistics.csv")
df.describe(include="all").to_csv(summary_file)

# -------------------------------
# Detect Important Columns
# -------------------------------
genre_col = next((c for c in df.columns if "genre" in c), None)
artist_col = next((c for c in df.columns if "artist" in c), None)
year_col = next((c for c in df.columns if "year" in c), None)
pop_col = next((c for c in df.columns if "popularity" in c), None)

numeric_cols = list(df.select_dtypes(include="number").columns)

images = []

# -------------------------------
# Chart 1
# -------------------------------
if genre_col and pop_col:
    plt.figure(figsize=(12,6))
    df.groupby(genre_col)[pop_col].mean().sort_values(ascending=False).plot(kind="bar")
    plt.title("Average Popularity by Genre")
    plt.tight_layout()
    file=os.path.join(OUTPUT_FOLDER,"genre_popularity.png")
    plt.savefig(file)
    images.append(file)
    plt.close()

# -------------------------------
# Chart 2
# -------------------------------
if artist_col and pop_col:
    plt.figure(figsize=(12,6))
    df.groupby(artist_col)[pop_col].mean().sort_values(ascending=False).head(10).plot(kind="bar")
    plt.title("Top 10 Artists")
    plt.tight_layout()
    file=os.path.join(OUTPUT_FOLDER,"top_artists.png")
    plt.savefig(file)
    images.append(file)
    plt.close()

# -------------------------------
# Chart 3
# -------------------------------
if year_col:
    plt.figure(figsize=(12,6))
    df.groupby(year_col).size().plot(kind="line",marker="o")
    plt.title("Songs Released per Year")
    plt.tight_layout()
    file=os.path.join(OUTPUT_FOLDER,"release_year.png")
    plt.savefig(file)
    images.append(file)
    plt.close()

# -------------------------------
# Histogram
# -------------------------------
if pop_col:
    plt.figure(figsize=(10,6))
    plt.hist(df[pop_col],bins=20)
    plt.title("Popularity Distribution")
    plt.tight_layout()
    file=os.path.join(OUTPUT_FOLDER,"popularity_histogram.png")
    plt.savefig(file)
    images.append(file)
    plt.close()

# -------------------------------
# Correlation Heatmap
# -------------------------------
if len(numeric_cols)>1:
    corr=df[numeric_cols].corr()

    plt.figure(figsize=(10,8))
    plt.imshow(corr,cmap="coolwarm",aspect="auto")
    plt.colorbar()

    plt.xticks(range(len(corr.columns)),corr.columns,rotation=90)
    plt.yticks(range(len(corr.columns)),corr.columns)

    plt.title("Correlation Matrix")
    plt.tight_layout()

    file=os.path.join(OUTPUT_FOLDER,"correlation.png")
    plt.savefig(file)
    images.append(file)
    plt.close()

# -------------------------------
# Boxplot
# -------------------------------
if pop_col:
    plt.figure(figsize=(8,6))
    plt.boxplot(df[pop_col])
    plt.title("Popularity Boxplot")
    plt.tight_layout()
    file=os.path.join(OUTPUT_FOLDER,"boxplot.png")
    plt.savefig(file)
    images.append(file)
    plt.close()

# -------------------------------
# Scatter Plot
# -------------------------------
if len(numeric_cols)>=2:
    plt.figure(figsize=(8,6))
    plt.scatter(df[numeric_cols[0]],df[numeric_cols[1]])
    plt.xlabel(numeric_cols[0])
    plt.ylabel(numeric_cols[1])
    plt.title("Scatter Plot")
    plt.tight_layout()
    file=os.path.join(OUTPUT_FOLDER,"scatter.png")
    plt.savefig(file)
    images.append(file)
    plt.close()

# -------------------------------
# PDF Report
# -------------------------------
pdf_file=os.path.join(OUTPUT_FOLDER,"Spotify_Report.pdf")

with PdfPages(pdf_file) as pdf:

    fig=plt.figure(figsize=(8.27,11.69))
    plt.axis("off")

    text=f"""
Spotify Dataset Report

Total Records : {len(df)}

Total Columns : {len(df.columns)}

Numeric Columns :
{', '.join(numeric_cols)}

Generated Automatically
"""

    plt.text(0.05,0.95,text,fontsize=14,va="top")
    pdf.savefig(fig)
    plt.close(fig)

    for img in images:
        fig=plt.figure(figsize=(10,6))
        plt.imshow(plt.imread(img))
        plt.axis("off")
        pdf.savefig(fig)
        plt.close(fig)

print("="*60)
print("SUCCESS")
print("="*60)
print("Cleaned CSV :",clean_file)
print("Summary CSV :",summary_file)
print("PDF Report  :",pdf_file)
print("Output Folder :",OUTPUT_FOLDER)
```
