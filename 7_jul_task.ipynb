```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# load data
df = pd.read_csv('/content/global-data-on-sustainable-energy (1).csv')
print(df.head())
```

```python
print("rows and columns:",df.shape)
print(df.info())
print(df.isnull().sum())
```

```python
df.groupby("Year")["Access to electricity (% of population)"].mean().plot(figsize=(10,5))
plt.title("Average Access to Electricity Over Years")
plt.xlabel("Year")
plt.ylabel("Electricity Access (%)")
plt.grid()
plt.show()
```

```python
df.groupby("Year")["Renewable energy share in the total final energy consumption (%)"].mean().plot(figsize=(10,5))
plt.title("Renewable Energy Share Over Years")
plt.xlabel("Year")
plt.ylabel("Renewable Energy (%)")
plt.grid()
plt.show()
```

```python
yearly_emissions = (
    df.groupby("Year")["Value_co2_emissions_kt_by_country"]
    .mean()
)

yearly_emissions.plot(figsize=(10, 5))

plt.title("Average CO2 Emissions Over Years")
plt.xlabel("Year")
plt.ylabel("Average CO2 Emissions (kt)")
plt.tight_layout()
plt.show()
```

```python
top = df.groupby("Entity")["Electricity from renewables (TWh)"].mean().sort_values(ascending=False).head(10)

top.plot(kind="bar", figsize=(10,5))
plt.title("Top 10 Countries by Renewable Electricity")
plt.xlabel("Country")
plt.ylabel("Renewable Electricity (TWh)")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 5))

sns.lineplot(
    data=df,
    x="Year",
    y="Value_co2_emissions_kt_by_country",
    estimator="mean",
    errorbar=None,
    marker="o"
)

plt.title("Average CO2 Emissions Over Years")
plt.xlabel("Year")
plt.ylabel("Average CO2 Emissions (kt)")
plt.tight_layout()
plt.show()
```

```python
print("Key Insights")
print("1. Access to electricity increased in many countries over time.")
print("2. Renewable energy share differs from country to country.")
print("3. CO2 emissions are higher in countries depending more on fossil fuels.")
print("4. Countries with better GDP usually have better electricity access.")
print("5. Renewable electricity generation is growing globally.")
```
