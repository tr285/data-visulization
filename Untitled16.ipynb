```python
# ================================================================
# PROFESSIONAL POWER BI-STYLE DASHBOARD + PDF EXPORT
# GOOGLE COLAB - RUN EVERYTHING IN ONE CELL
# ================================================================

!pip install -q pandas numpy matplotlib reportlab

import os
import textwrap
import warnings
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.patches import FancyBboxPatch
from matplotlib.backends.backend_pdf import PdfPages
from google.colab import files
from IPython.display import display

warnings.filterwarnings("ignore")


# ================================================================
# 1. UPLOAD DATASET
# ================================================================

print("Upload your mental_health_tech_survey.csv file")

uploaded = files.upload()

if not uploaded:
    raise ValueError("No CSV file uploaded.")

file_name = next(iter(uploaded.keys()))


# ================================================================
# 2. LOAD DATASET
# ================================================================

try:
    df_raw = pd.read_csv(file_name)

except UnicodeDecodeError:
    df_raw = pd.read_csv(file_name, encoding="latin1")


df = df_raw.copy()


# ================================================================
# 3. CLEAN COLUMN NAMES
# ================================================================

df.columns = (
    df.columns
    .astype(str)
    .str.strip()
    .str.lower()
    .str.replace(r"[^a-z0-9]+", "_", regex=True)
    .str.strip("_")
)


# ================================================================
# 4. REMOVE DUPLICATES
# ================================================================

original_rows = len(df)

df = df.drop_duplicates().reset_index(drop=True)

duplicates_removed = original_rows - len(df)


# ================================================================
# 5. CLEAN AGE
# ================================================================

invalid_ages = 0

if "age" in df.columns:

    df["age"] = pd.to_numeric(df["age"], errors="coerce")

    invalid_mask = (
        df["age"].isna()
        | (df["age"] < 18)
        | (df["age"] > 80)
    )

    invalid_ages = int(invalid_mask.sum())

    df.loc[invalid_mask, "age"] = np.nan

    if df["age"].notna().any():
        df["age"] = df["age"].fillna(df["age"].median())


# ================================================================
# 6. CLEAN GENDER
# ================================================================

def standardize_gender(value):

    if pd.isna(value):
        return "Other"

    value = str(value).strip().lower()

    male_terms = {
        "male", "m", "man", "cis male", "male (cis)",
        "male-ish", "maile", "mal", "make", "msle",
        "mail", "malr", "cis man", "guy (-ish) ^_^"
    }

    female_terms = {
        "female", "f", "woman", "cis female",
        "female (cis)", "femake", "woman-identified",
        "cis-female/femme", "female woman"
    }

    if value in male_terms:
        return "Male"

    if value in female_terms:
        return "Female"

    return "Other"


if "gender" in df.columns:
    df["gender"] = df["gender"].apply(standardize_gender)


# ================================================================
# 7. CLEAN TEXT COLUMNS
# ================================================================

text_columns = df.select_dtypes(include=["object", "string"]).columns

for column in text_columns:

    df[column] = (
        df[column]
        .astype("string")
        .str.strip()
        .replace({
            "": pd.NA,
            "nan": pd.NA,
            "NaN": pd.NA,
            "None": pd.NA,
            "null": pd.NA
        })
        .fillna("Unknown")
    )


# ================================================================
# 8. STANDARDIZE YES / NO COLUMNS
# ================================================================

binary_columns = [
    "self_employed",
    "family_history",
    "treatment",
    "remote_work",
    "tech_company",
    "obs_consequence"
]

for column in binary_columns:

    if column in df.columns:

        df[column] = (
            df[column]
            .astype(str)
            .str.strip()
            .str.lower()
            .replace({
                "yes": "Yes",
                "y": "Yes",
                "true": "Yes",
                "1": "Yes",
                "no": "No",
                "n": "No",
                "false": "No",
                "0": "No"
            })
        )


# ================================================================
# 9. DATA QUALITY METRICS
# ================================================================

missing_before = int(df_raw.isnull().sum().sum())
missing_after = int(df.isnull().sum().sum())


# ================================================================
# 10. SAVE CLEANED DATA
# ================================================================

cleaned_file = "cleaned_mental_health_tech_survey.csv"

df.to_csv(cleaned_file, index=False)


# ================================================================
# 11. KPI CALCULATIONS
# ================================================================

total_respondents = len(df)

average_age = (
    df["age"].mean()
    if "age" in df.columns
    else 0
)

treatment_rate = (
    df["treatment"].astype(str).str.lower().eq("yes").mean() * 100
    if "treatment" in df.columns
    else 0
)

remote_rate = (
    df["remote_work"].astype(str).str.lower().eq("yes").mean() * 100
    if "remote_work" in df.columns
    else 0
)

country_count = (
    df["country"].nunique()
    if "country" in df.columns
    else 0
)


# ================================================================
# 12. ADDITIONAL ANALYTICAL METRICS
# ================================================================

family_history_rate = (
    df["family_history"].astype(str).str.lower().eq("yes").mean() * 100
    if "family_history" in df.columns
    else 0
)

benefits_yes_rate = (
    df["benefits"].astype(str).str.lower().eq("yes").mean() * 100
    if "benefits" in df.columns
    else 0
)

top_country = (
    df["country"].value_counts().idxmax()
    if "country" in df.columns and not df.empty
    else "N/A"
)

top_country_count = (
    int(df["country"].value_counts().max())
    if "country" in df.columns and not df.empty
    else 0
)


# ================================================================
# 13. PREPARE CHART DATA
# ================================================================

gender_data = (
    df["gender"].value_counts()
    if "gender" in df.columns
    else pd.Series(dtype=int)
)

country_data = (
    df["country"]
    .value_counts()
    .head(10)
    .sort_values()
    if "country" in df.columns
    else pd.Series(dtype=int)
)

treatment_data = (
    df["treatment"].value_counts()
    if "treatment" in df.columns
    else pd.Series(dtype=int)
)

benefits_data = (
    df["benefits"].value_counts()
    if "benefits" in df.columns
    else pd.Series(dtype=int)
)

family_treatment = None

if {"family_history", "treatment"}.issubset(df.columns):

    family_treatment = pd.crosstab(
        df["family_history"],
        df["treatment"]
    )


# ================================================================
# 14. AUTOMATIC INSIGHTS
# ================================================================

insights = []

insights.append(
    f"{treatment_rate:.1f}% of respondents reported seeking "
    f"mental health treatment."
)

insights.append(
    f"{family_history_rate:.1f}% reported a family history "
    f"of mental health conditions."
)

insights.append(
    f"{remote_rate:.1f}% of respondents reported working remotely."
)

insights.append(
    f"{top_country} contributed the largest respondent group "
    f"with {top_country_count:,} survey records."
)

insights.append(
    f"{benefits_yes_rate:.1f}% reported that their employer "
    f"provides mental health benefits."
)


# ================================================================
# 15. DASHBOARD CONFIGURATION
# ================================================================

plt.rcParams.update({
    "font.size": 9,
    "axes.titlesize": 12,
    "axes.titleweight": "bold",
    "axes.labelsize": 9,
    "xtick.labelsize": 8,
    "ytick.labelsize": 8
})

fig = plt.figure(figsize=(16, 11.2))

fig.patch.set_facecolor("#F3F4F6")


# ================================================================
# 16. HEADER
# ================================================================

header = FancyBboxPatch(
    (0.025, 0.91),
    0.95,
    0.07,
    boxstyle="round,pad=0.005,rounding_size=0.008",
    transform=fig.transFigure,
    facecolor="#172554",
    edgecolor="none"
)

fig.add_artist(header)

fig.text(
    0.05,
    0.953,
    "MENTAL HEALTH IN TECHNOLOGY",
    fontsize=22,
    fontweight="bold",
    color="white",
    va="center"
)

fig.text(
    0.05,
    0.925,
    "Workforce Mental Health Analytics | Executive Dashboard",
    fontsize=10,
    color="#CBD5E1"
)

fig.text(
    0.95,
    0.945,
    "ANALYTICS REPORT",
    fontsize=10,
    fontweight="bold",
    color="white",
    ha="right"
)


# ================================================================
# 17. KPI CARD FUNCTION
# ================================================================

def add_kpi_card(x, y, width, height, title, value, note):

    card = FancyBboxPatch(
        (x, y),
        width,
        height,
        boxstyle="round,pad=0.005,rounding_size=0.008",
        transform=fig.transFigure,
        facecolor="white",
        edgecolor="#D1D5DB",
        linewidth=0.8
    )

    fig.add_artist(card)

    fig.text(
        x + 0.012,
        y + height - 0.023,
        title.upper(),
        fontsize=8,
        fontweight="bold",
        color="#64748B"
    )

    fig.text(
        x + 0.012,
        y + 0.035,
        value,
        fontsize=20,
        fontweight="bold",
        color="#0F172A"
    )

    fig.text(
        x + 0.012,
        y + 0.012,
        note,
        fontsize=7.5,
        color="#64748B"
    )


# ================================================================
# 18. KPI CARDS
# ================================================================

card_y = 0.81
card_h = 0.075
card_w = 0.176
card_gap = 0.014

kpi_cards = [
    (
        "Respondents",
        f"{total_respondents:,}",
        f"{duplicates_removed} duplicate rows removed"
    ),
    (
        "Average Age",
        f"{average_age:.1f}",
        f"{invalid_ages} invalid ages corrected"
    ),
    (
        "Treatment Rate",
        f"{treatment_rate:.1f}%",
        "Share reporting treatment"
    ),
    (
        "Remote Work",
        f"{remote_rate:.1f}%",
        "Share working remotely"
    ),
    (
        "Countries",
        f"{country_count:,}",
        f"Top market: {top_country}"
    )
]

for i, (title, value, note) in enumerate(kpi_cards):

    add_kpi_card(
        0.025 + i * (card_w + card_gap),
        card_y,
        card_w,
        card_h,
        title,
        value,
        note
    )


# ================================================================
# 19. PANEL HELPER
# ================================================================

def create_panel(position, title):

    ax = fig.add_axes(position)

    ax.set_facecolor("white")

    for spine in ax.spines.values():
        spine.set_color("#D1D5DB")
        spine.set_linewidth(0.8)

    ax.set_title(
        title,
        loc="left",
        pad=10,
        color="#0F172A"
    )

    return ax


# ================================================================
# 20. GRAPH 1 - GENDER DISTRIBUTION
# ================================================================

ax1 = create_panel(
    [0.035, 0.55, 0.28, 0.22],
    "Gender Distribution"
)

if not gender_data.empty:

    wedges, texts, autotexts = ax1.pie(
        gender_data.values,
        labels=gender_data.index,
        autopct="%1.1f%%",
        startangle=90,
        pctdistance=0.78,
        wedgeprops={
            "width": 0.42,
            "edgecolor": "white"
        }
    )

    ax1.text(
        0,
        0.03,
        f"{gender_data.sum():,}",
        ha="center",
        va="center",
        fontsize=16,
        fontweight="bold"
    )

    ax1.text(
        0,
        -0.12,
        "Respondents",
        ha="center",
        fontsize=8,
        color="#64748B"
    )


# ================================================================
# 21. GRAPH 2 - AGE DISTRIBUTION
# ================================================================

ax2 = create_panel(
    [0.345, 0.55, 0.30, 0.22],
    "Respondent Age Distribution"
)

if "age" in df.columns:

    ax2.hist(
        df["age"].dropna(),
        bins=20,
        edgecolor="white"
    )

    ax2.axvline(
        average_age,
        linestyle="--",
        linewidth=1.5,
        label=f"Average: {average_age:.1f}"
    )

    ax2.set_xlabel("Age")
    ax2.set_ylabel("Respondents")
    ax2.legend(frameon=False, fontsize=8)
    ax2.grid(axis="y", alpha=0.2)


# ================================================================
# 22. GRAPH 3 - TREATMENT SEEKING
# ================================================================

ax3 = create_panel(
    [0.665, 0.55, 0.30, 0.22],
    "Treatment Seeking Behaviour"
)

if not treatment_data.empty:

    bars = ax3.bar(
        treatment_data.index.astype(str),
        treatment_data.values
    )

    ax3.set_ylabel("Respondents")
    ax3.grid(axis="y", alpha=0.2)

    for bar in bars:

        ax3.text(
            bar.get_x() + bar.get_width() / 2,
            bar.get_height(),
            f"{int(bar.get_height()):,}",
            ha="center",
            va="bottom",
            fontsize=8,
            fontweight="bold"
        )


# ================================================================
# 23. GRAPH 4 - TOP COUNTRIES
# ================================================================

ax4 = create_panel(
    [0.035, 0.265, 0.36, 0.235],
    "Top 10 Countries by Survey Participation"
)

if not country_data.empty:

    bars = ax4.barh(
        country_data.index.astype(str),
        country_data.values
    )

    ax4.set_xlabel("Respondents")
    ax4.grid(axis="x", alpha=0.2)

    for bar in bars:

        ax4.text(
            bar.get_width(),
            bar.get_y() + bar.get_height() / 2,
            f" {int(bar.get_width()):,}",
            va="center",
            fontsize=7.5
        )


# ================================================================
# 24. GRAPH 5 - FAMILY HISTORY VS TREATMENT
# ================================================================

ax5 = create_panel(
    [0.425, 0.265, 0.30, 0.235],
    "Family History vs Treatment"
)

if family_treatment is not None and not family_treatment.empty:

    family_treatment.plot(
        kind="bar",
        ax=ax5,
        width=0.72
    )

    ax5.set_xlabel("Family History")
    ax5.set_ylabel("Respondents")
    ax5.tick_params(axis="x", rotation=0)
    ax5.legend(
        title="Treatment",
        frameon=False,
        fontsize=7
    )
    ax5.grid(axis="y", alpha=0.2)


# ================================================================
# 25. GRAPH 6 - EMPLOYER BENEFITS
# ================================================================

ax6 = create_panel(
    [0.755, 0.265, 0.21, 0.235],
    "Mental Health Benefits"
)

if not benefits_data.empty:

    wedges, texts, autotexts = ax6.pie(
        benefits_data.values,
        labels=benefits_data.index.astype(str),
        autopct="%1.0f%%",
        startangle=90,
        textprops={"fontsize": 7}
    )


# ================================================================
# 26. KEY INSIGHTS PANEL
# ================================================================

insight_ax = create_panel(
    [0.035, 0.045, 0.57, 0.17],
    "Key Business Insights"
)

insight_ax.set_xticks([])
insight_ax.set_yticks([])

y_position = 0.78

for number, insight in enumerate(insights, start=1):

    wrapped_text = textwrap.fill(
        insight,
        width=72
    )

    insight_ax.text(
        0.03,
        y_position,
        f"{number}. {wrapped_text}",
        transform=insight_ax.transAxes,
        fontsize=8.5,
        color="#334155",
        va="top"
    )

    y_position -= 0.17


# ================================================================
# 27. DATA QUALITY PANEL
# ================================================================

quality_ax = create_panel(
    [0.635, 0.045, 0.33, 0.17],
    "Data Quality & Governance"
)

quality_ax.set_xticks([])
quality_ax.set_yticks([])

quality_items = [
    ("Original Records", f"{len(df_raw):,}"),
    ("Cleaned Records", f"{len(df):,}"),
    ("Duplicates Removed", f"{duplicates_removed:,}"),
    ("Invalid Ages Corrected", f"{invalid_ages:,}"),
    ("Missing Before Cleaning", f"{missing_before:,}"),
    ("Missing After Cleaning", f"{missing_after:,}")
]

quality_y = 0.78

for label, value in quality_items:

    quality_ax.text(
        0.05,
        quality_y,
        label,
        transform=quality_ax.transAxes,
        fontsize=8.5,
        color="#64748B"
    )

    quality_ax.text(
        0.95,
        quality_y,
        value,
        transform=quality_ax.transAxes,
        fontsize=9,
        fontweight="bold",
        color="#0F172A",
        ha="right"
    )

    quality_y -= 0.14


# ================================================================
# 28. FOOTER
# ================================================================

fig.text(
    0.035,
    0.015,
    "Source: Mental Health in Technology Survey | "
    "Prepared with Python Analytics Dashboard",
    fontsize=7.5,
    color="#64748B"
)

fig.text(
    0.965,
    0.015,
    "Executive Analytics Report",
    fontsize=7.5,
    color="#64748B",
    ha="right"
)


# ================================================================
# 29. DISPLAY DASHBOARD IN COLAB
# ================================================================

display(fig)


# ================================================================
# 30. EXPORT HIGH-RESOLUTION PDF
# ================================================================

pdf_file = "mental_health_powerbi_dashboard.pdf"

with PdfPages(pdf_file) as pdf:

    pdf.savefig(
        fig,
        bbox_inches="tight",
        pad_inches=0.15
    )


# ================================================================
# 31. EXPORT HIGH-RESOLUTION PNG
# ================================================================

png_file = "mental_health_powerbi_dashboard.png"

fig.savefig(
    png_file,
    dpi=300,
    bbox_inches="tight",
    pad_inches=0.15
)


# ================================================================
# 32. FINAL SUMMARY
# ================================================================

print("\n" + "=" * 70)
print("PROFESSIONAL DASHBOARD CREATED SUCCESSFULLY")
print("=" * 70)

print("Dashboard Graphs: 6")
print("KPI Cards: 5")
print("Calculated Insights:", len(insights))
print("Cleaned Dataset:", cleaned_file)
print("PDF Dashboard:", pdf_file)
print("PNG Dashboard:", png_file)

print("=" * 70)


# ================================================================
# 33. DOWNLOAD FILES
# ================================================================

files.download(pdf_file)
files.download(png_file)
files.download(cleaned_file)
```
