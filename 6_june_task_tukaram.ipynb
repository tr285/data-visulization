```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("uptoskills/task 1/3b01bcb8-0b14-4abf-b6f2-c1bfd384ba69.csv")

pollutant_analysis = df.groupby('pollutant_id')['pollutant_avg'].mean().sort_values(ascending=False)
state_analysis = df.groupby('state')['pollutant_avg'].mean().sort_values(ascending=False)
city_analysis = df.groupby('city')['pollutant_avg'].mean().sort_values(ascending=False)

print("Dataset Shape:", df.shape)
print("States:", df['state'].nunique())
print("Cities:", df['city'].nunique())

print("\nAverage Pollution by Pollutant")
print(pollutant_analysis)

print("\nTop 10 Polluted States")
print(state_analysis.head(10))

print("\nTop 10 Polluted Cities")
print(city_analysis.head(10))

plt.figure(figsize=(8,5))
pollutant_analysis.plot(kind='bar')
plt.title("Average Pollution by Pollutant")
plt.tight_layout()
plt.show()

plt.figure(figsize=(10,5))
state_analysis.head(10).plot(kind='bar')
plt.title("Top 10 Polluted States")
plt.tight_layout()
plt.show()

plt.figure(figsize=(10,5))
city_analysis.head(10).plot(kind='bar')
plt.title("Top 10 Polluted Cities")
plt.tight_layout()
plt.show()
```
