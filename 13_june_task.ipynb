```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("globalterrorismdb_0718dist.csv", encoding="latin1", low_memory=False)

print("Shape:", df.shape)

print("\nDuplicates:", df.duplicated().sum())

print("\nMissing Values:")
print(df.isnull().sum().sort_values(ascending=False).head(20))

print("\nTotal Attacks:", len(df))

print("\nTop 10 Countries:")
print(df['country_txt'].value_counts().head(10))

print("\nTop 10 Regions:")
print(df['region_txt'].value_counts().head(10))

print("\nTop Attack Types:")
print(df['attacktype1_txt'].value_counts().head(10))

print("\nTop Targets:")
print(df['targtype1_txt'].value_counts().head(10))

print("\nTop Terrorist Groups:")
print(df['gname'].value_counts().head(10))

print("\nTotal Fatalities:", df['nkill'].fillna(0).sum())
print("Total Injuries:", df['nwound'].fillna(0).sum())

yearly = df.groupby('iyear').size()

plt.figure(figsize=(12,5))
yearly.plot()
plt.title("Terrorist Attacks by Year")
plt.xlabel("Year")
plt.ylabel("Number of Attacks")
plt.show()

country = df['country_txt'].value_counts().head(10)

plt.figure(figsize=(10,5))
country.plot(kind='bar')
plt.title("Top 10 Countries by Attacks")
plt.show()

attack = df['attacktype1_txt'].value_counts().head(10)

plt.figure(figsize=(10,5))
attack.plot(kind='bar')
plt.title("Top Attack Types")
plt.show()

group = df['gname'].value_counts().head(10)

plt.figure(figsize=(10,5))
group.plot(kind='bar')
plt.title("Top Terrorist Groups")
plt.show()
```
