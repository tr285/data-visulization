```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
data = pd.read_csv("historical_emissions 3.csv")

data = data.drop_duplicates()

years = [str(y) for y in range(1990, 2019)]

top10 = data[["Country", "2018"]].copy()
top10["2018"] = pd.to_numeric(top10["2018"], errors="coerce")
top10 = top10.dropna()
top10 = top10.sort_values("2018", ascending=False).head(10)

global_emission = data[years].sum()

fig, axes = plt.subplots(2, 4, figsize=(22, 10))

# 1 Top Countries
axes[0,0].bar(top10["Country"], top10["2018"])
axes[0,0].set_title("Top 10 Emitters")
axes[0,0].tick_params(axis='x', rotation=90)

# 2 Sector Pie
if "Sector" in data.columns:
    sector_count = data["Sector"].value_counts()
    axes[0,1].pie(sector_count,
                  labels=sector_count.index,
                  autopct="%1.1f%%")
    axes[0,1].set_title("Sector Distribution")

# 3 Sector Count
if "Sector" in data.columns:
    sector_count.plot(kind="bar", ax=axes[0,2])
    axes[0,2].set_title("Sector Count")

# 4 Country Distribution
axes[0,3].hist(top10["2018"], bins=10)
axes[0,3].set_title("Country Distribution")

# 5 Gas Pie
if "Gas" in data.columns:
    gas_count = data["Gas"].value_counts()
    axes[1,0].pie(gas_count,
                  labels=gas_count.index,
                  autopct="%1.1f%%")
    axes[1,0].set_title("Gas Distribution")

# 6 Heatmap
heatmap_data = top10.set_index("Country")[["2018"]]

sns.heatmap(
    heatmap_data,
    annot=True,
    cmap="Reds",
    ax=axes[1,1]
)

axes[1,1].set_title("Heatmap")

# 7 Global Trend
axes[1,2].plot(years, global_emission)
axes[1,2].set_title("Global Trend")
axes[1,2].tick_params(axis='x', rotation=90)

# 8 China USA India
countries = ["China", "United States", "India"]

for country in countries:

    cdata = data[data["Country"] == country]

    if not cdata.empty:
        axes[1,3].plot(
            years,
            cdata[years].iloc[0],
            label=country
        )

axes[1,3].legend()
axes[1,3].set_title("China vs USA vs India")
axes[1,3].tick_params(axis='x', rotation=90)

plt.suptitle(
    "Carbon Emission Trends, Sources & Country Insights Dashboard",
    fontsize=18
)

plt.tight_layout()

plt.savefig(
    "carbon_dashboard.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```
