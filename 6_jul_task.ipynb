```python
import pandas as pd
df =pd.read_csv('/content/flight_data_2024.csv')
display(df.head())
```

```python
df.info()
```

```python
display(df.describe())
```

```python
display(df.isnull().sum())
```

```python
df['fl_date'] = pd.to_datetime(df['fl_date'])
df.info()
```

```python
columns_to_fill = ['dep_time', 'dep_delay', 'taxi_out', 'wheels_off', 'wheels_on',
                     'taxi_in', 'arr_time', 'arr_delay', 'actual_elapsed_time', 'air_time']

for col in columns_to_fill:
    if df[col].isnull().any():
        median_val = df[col].median()
        df[col] = df[col].fillna(median_val)
        print(f"Filled missing values in '{col}' with median: {median_val}")
df['cancellation_code'] = df['cancellation_code'].fillna('Not Cancelled')
print("Filled missing values in 'cancellation_code' with 'Not Cancelled'")
display(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
daily_avg_dep_delay = df.groupby('fl_date')['dep_delay'].mean().reset_index()
daily_avg_dep_delay = daily_avg_dep_delay.sort_values(by='fl_date')
plt.figure(figsize=(14, 7))
sns.lineplot(data=daily_avg_dep_delay, x='fl_date', y='dep_delay')
plt.title('Average Daily Departure Delay')
plt.xlabel('Date')
plt.ylabel('Average Departure Delay (minutes)')
plt.grid(True)
plt.tight_layout()
plt.show()
display(daily_avg_dep_delay.head())
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['dep_delay'], bins=50, kde=True)
plt.title('Distribution of Departure Delays')
plt.xlabel('Departure Delay (minutes)')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
avg_delay_by_carrier = df.groupby('op_unique_carrier')['dep_delay'].mean().sort_values(ascending=False)

plt.figure(figsize=(14, 7))
sns.barplot(x=avg_delay_by_carrier.index, y=avg_delay_by_carrier.values, palette='viridis')
plt.title('Average Departure Delay by Operating Carrier')
plt.xlabel('Operating Carrier')
plt.ylabel('Average Departure Delay (minutes)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
avg_delay_by_origin = df.groupby('origin_city_name')['dep_delay'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(14, 7))
sns.barplot(x=avg_delay_by_origin.index, y=avg_delay_by_origin.values, palette='plasma')
plt.title('Top 10 Origin Cities by Average Departure Delay')
plt.xlabel('Origin City')
plt.ylabel('Average Departure Delay (minutes)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python

numerical_df = df.select_dtypes(include=['float64', 'int64'])
corr_matrix = numerical_df.corr()

plt.figure(figsize=(18, 16))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python

cancellation_counts = df['cancellation_code'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(cancellation_counts, labels=cancellation_counts.index, autopct='%1.1f%%', startangle=140, wedgeprops=dict(width=0.3))
plt.title('Distribution of Cancellation Reasons')
plt.ylabel('')
plt.show()
```

```python
delay_types = ['carrier_delay', 'weather_delay', 'nas_delay', 'security_delay', 'late_aircraft_delay']
average_delay_contributions = df[delay_types].mean()


fig, ax = plt.subplots(figsize=(10, 6))

cumulative_sum = 0
left_coords = []
heights = []
labels = []
for i, (delay_type, avg_contribution) in enumerate(average_delay_contributions.items()):
    left_coords.append(i)
    heights.append(avg_contribution)
    labels.append(delay_type.replace('_delay', '').title())
    ax.bar(i, avg_contribution, bottom=cumulative_sum, color=plt.cm.viridis(i/len(delay_types)))
    cumulative_sum += avg_contribution
left_coords.append(len(delay_types))
heights.append(cumulative_sum)
labels.append('Total Avg Delay')
ax.bar(len(delay_types), cumulative_sum, color='gray')

ax.set_xticks(range(len(labels)))
ax.set_xticklabels(labels, rotation=45, ha='right')
ax.set_ylabel('Average Delay Contribution (minutes)')
ax.set_title('Average Delay Contribution by Type')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
origin_flight_counts = df['origin_city_name'].value_counts().head(20)

plt.figure(figsize=(12, 7))
sns.barplot(x=origin_flight_counts.index, y=origin_flight_counts.values, palette='coolwarm')
plt.title('Top 20 Origin Cities by Flight Count')
plt.xlabel('Origin City')
plt.ylabel('Number of Flights')
plt.xticks(rotation=60, ha='right')
plt.tight_layout()
plt.show()

print("To plot these on a geographical map, we would need latitude and longitude coordinates for each city.")
```
