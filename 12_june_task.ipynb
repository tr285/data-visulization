```python
import pandas as pd
import numpy as np

df = pd.read_csv("Bengaluru_House_Data.csv")

print("\nDATASET SHAPE")
print(df.shape)

print("\nCOLUMN NAMES")
print(df.columns)

print("\nFIRST 5 ROWS")
print(df.head())

print("\nMISSING VALUES")
print(df.isnull().sum())

print("\nDUPLICATE ROWS")
print(df.duplicated().sum())

print("\nDATA TYPES")
print(df.dtypes)

print("\nNUMERICAL SUMMARY")
print(df.describe())

print("\nTOP 10 LOCATIONS")
if 'location' in df.columns:
    print(df['location'].value_counts().head(10))

if 'price' in df.columns:
    print("\nPRICE STATISTICS")
    print("Average Price:", round(df['price'].mean(), 2))
    print("Maximum Price:", df['price'].max())
    print("Minimum Price:", df['price'].min())

if 'size' in df.columns:
    print("\nTOP PROPERTY TYPES")
    print(df['size'].value_counts().head(10))

if 'bath' in df.columns:
    print("\nAVERAGE BATHROOMS:", round(df['bath'].mean(), 2))

if 'balcony' in df.columns:
    print("AVERAGE BALCONIES:", round(df['balcony'].mean(), 2))

print("\nCORRELATION MATRIX")
numeric_df = df.select_dtypes(include=np.number)
print(numeric_df.corr())

insights = []

if 'price' in df.columns:
    insights.append(f"Average property price is {round(df['price'].mean(),2)}.")
    insights.append(f"Highest property price is {df['price'].max()}.")
    insights.append(f"Lowest property price is {df['price'].min()}.")

if 'location' in df.columns:
    top_location = df['location'].value_counts().idxmax()
    insights.append(f"Most properties are located in {top_location}.")

if 'size' in df.columns:
    top_size = df['size'].value_counts().idxmax()
    insights.append(f"Most common property type is {top_size}.")

print("\nKEY INSIGHTS")
for i, insight in enumerate(insights, 1):
    print(f"{i}. {insight}")
```
